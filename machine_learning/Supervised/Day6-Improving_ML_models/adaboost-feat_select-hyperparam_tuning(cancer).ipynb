# Running Models

- Use Adaboost on the ‘Breast Cancer’ dataset to predict a diagnosis:
- Clean and preprocess data.
- Run the model on the data.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectFromModel
```

```python
bc_data = pd.read_csv('data-breast-cancer.csv')
```

```python
bc_data.head()
```

```python
bc_data.describe()
```

```python
bc_data.columns
```

```python
# Isolate independent and dependent variables
bc_y = bc_data['diagnosis']
bc_x = bc_data.drop(['diagnosis', 'id', 'Unnamed: 32'], axis=1)
```

```python
# Assign independent and dependent variables
X = bc_x
y = bc_y  # Variable: Diagnosis
```

```python
# Split data train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Adaboost

```python
# Checking distribution
bc_data['diagnosis'].value_counts()
```

```python
# Build Adaboost model
ada = AdaBoostClassifier(n_estimators=100)
ada.fit(X_train,y_train)
```

```python
# Evaluating the model
scores_f = cross_val_score(ada, X_train, y_train, cv=3, scoring='roc_auc')
scores_f.mean()
```

# Feature Selection

- Create plots to understand the data further:
- Are there any features that can be removed (or added)?
- Run the model again, and compare results from part 1.

```python
# Checking for feature importance
f_imp = ada.feature_importances_
f_names = list(X_train)
```

```python
plt.scatter(range(1, len(f_imp)+1), f_imp)
plt.xticks(range(1, len(f_imp)+1), f_names, rotation=90)
plt.ylabel('Feature Importance')
plt.show()
```

```python
# Putting feature importance in a data frame
f_imp_df = pd.DataFrame(f_imp, f_names)
f_imp_df = f_imp_df.rename(columns={0: 'Feature Importance'})
f_imp_df = f_imp_df.sort_values('Feature Importance', ascending=False)
f_imp_df.head()
```

```python
# Checking correlations
corr = X.corr()
corr.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

```python
# Create a selector object to identify features with importance > 0.05
sfm = SelectFromModel(ada, threshold=0.05).fit(X_train, y_train)
```

```python
# Transform the data to create a new dataset containing only the most important features
# Note: We have to apply the transform to both the training X and test X data.
X_train_new = sfm.transform(X_train)
```

```python
# Create a new random forest classifier for the most important features
ada_new = AdaBoostClassifier(n_estimators=100)

# Train the new classifier on the new dataset containing the most important features
ada_new.fit(X_train_new, y_train)
```

```python
# Evaluating the model
scores_f = cross_val_score(ada_new, X_train_new, y_train, cv=3, scoring='roc_auc')
scores_f.mean()
```

# Hyperparameter Tuning

- Using Grid Search and Random Search, tune the hyperparameters of the three models:
- Run both Grid Search and Random Search
- Compare with previous results in part 1 and 2.

```python
param_grid_random = {
    'n_estimators': [250, 500, 1000],
}
```

```python
# Build Randomized Search model
ada_rs = RandomizedSearchCV(ada_new, param_grid_random, n_iter=100, cv=3, scoring='roc_auc', n_jobs=-1)
ada_rs.fit(X_train_new, y_train)
```

```python
# Checking for best parameters
ada_rs.best_params_
```

```python
# Best Model Randomized Search
# Applying best model
ada_rs.best_estimator_
```

```python
# Evaluate Randomized Search model
scores_rs = cross_val_score(ada_rs.best_estimator_, X_train_new, y_train, cv=3, scoring='roc_auc')
scores_rs.mean()
```

```python
param_grid_gs = {
    'n_estimators': [500, 750, 1000],
}
```

```python
# Build Grid Search model
ada_gs = GridSearchCV(ada_new, param_grid_gs, cv=3, scoring='roc_auc', n_jobs=-1)
ada_gs.fit(X_train_new, y_train)
```

```python
# Checking for best parameters
ada_gs.best_params_
```

```python
# Best Model Grid Search
# Applying best model
ada_gs.best_estimator_
```

```python
# Evaluate Grid Search model
scores_gs = cross_val_score(ada_gs.best_estimator_, X_train_new, y_train, cv=3, scoring='roc_auc')
scores_gs.mean()
```
