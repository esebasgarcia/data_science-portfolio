# Part 2: Ridge Lasso Regression

Perform Lasso and Ridge regression on wine dataset. Compare the value of coefficients obtained by these two methods with standard regression coefficients.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
from sklearn import metrics
```

```python
plt.rcParams["figure.figsize"] = (8, 8)
plt.rcParams["font.size"] = 14
```

```python
wine = pd.read_csv('winequality-white.csv', delimiter=';')
wine.pop('quality')
wine.head()
```

```python
X = wine.iloc[:,:-1]
y = wine.iloc[:,10].values
```

```python
# Normalise Data
scaler = StandardScaler()
xs_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
xs_scaled.head()
```

```python
plt.plot(y, X, '.')
plt.xlabel('Alcohol (y)')
plt.ylabel('Variables (X)')
```

```python
X_train, X_test, y_train, y_test = train_test_split(xs_scaled, y,
                                                    test_size=.33,
                                                    random_state=0)

# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
ridge = Ridge(alpha=0)
ridge.fit(X_train, y_train)
predictions = ridge.predict(X_test)
pd.DataFrame(ridge.coef_, wine.columns.values[0:10],
             columns=['Coefficients'])
```

```python
#Calculate the metrics
print('MAE', metrics.mean_absolute_error(y_test, predictions))
print('MSE', metrics.mean_squared_error(y_test, predictions))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('Explained Var Score', metrics.explained_variance_score(y_test, predictions))
```

## Ridge Regression

```python
# Logarithmic

alphas_log = 10**np.linspace(-10, 10, 101)*0.5

ridge = Ridge()
coefs = []
mse = []
rscore = []

for a in alphas_log:
    ridge.set_params(alpha=a)
    ridge.fit(X_train, y_train)
    predictions = ridge.predict(X_test)
    coefs.append(ridge.coef_)
    mse.append(metrics.mean_squared_error(y_test, predictions))
    rscore.append(ridge.score(X_test, y_test))

ax = plt.gca()
ax.plot(alphas_log, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
```

```python
df = pd.DataFrame(coefs)
df.columns = wine.iloc[:, :-1].columns.values
df2 = pd.DataFrame(alphas_log)
df2.columns = ['alpha']
df3 = pd.DataFrame(mse)
df3.columns = ['mse']
df4 = pd.DataFrame(rscore)
df4.columns = ['score']
ridge_df = df2.join(df3)
ridge_df = ridge_df.join(df4)
ridge_df = ridge_df.join(df)
```

```python
ridge_df.head()
```

```python
ridge_df.loc[ridge_df['mse'] == min(ridge_df['mse'])]
```

```python
g = sns.pairplot(ridge_df, x_vars=['alpha'],
                 y_vars=['mse', 'score', 'fixed acidity', 'volatile acidity', 'citric acid',
                         'residual sugar', 'chlorides', 'free sulfur dioxide',
                         'total sulfur dioxide', 'density', 'pH', 'sulphates'])
```

## Lasso Regression

```python
# Logarithmic

alphas_log = 10**np.linspace(-10, 10, 101)*0.5

lasso = Lasso()
coefs = []
mse = []
rscore = []

for a in alphas_log:
    lasso.set_params(alpha=a)
    lasso.fit(X_train, y_train)
    predictions = lasso.predict(X_test)
    coefs.append(lasso.coef_)
    mse.append(metrics.mean_squared_error(y_test, predictions))
    rscore.append(lasso.score(X_test, y_test))

ax = plt.gca()
ax.plot(alphas_log, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
```

```python
df = pd.DataFrame(coefs)
df.columns = wine.iloc[:, :-1].columns.values
df2 = pd.DataFrame(alphas_log)
df2.columns = ['alpha']
df3 = pd.DataFrame(mse)
df3.columns = ['mse']
df4 = pd.DataFrame(rscore)
df4.columns = ['score']
lasso_df = df2.join(df3)
lasso_df = lasso_df.join(df4)
lasso_df = lasso_df.join(df)
```
